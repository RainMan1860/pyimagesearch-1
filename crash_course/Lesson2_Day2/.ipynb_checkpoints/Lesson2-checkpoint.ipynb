# OpenCV Tutorial: A Guide to Learn OpenCV

We will learn the basics of OpenCV and image processing. Get down the fundamentals. Stary with basic functions and end with counting objects in images  using contours

```python
import cv2
import imutils
import os
from matplotlib import pyplot as plt
import numpy as np
```

```python
def display_BGR(image):
    ''' OpenCV stores in BGR rather than RGB. This will convert a BGR image
    into an RGB image and display it
    '''
    im2 = image.copy()
    im2[:, :, 0] = image[:, :, 2]
    im2[:, :, 2] = image[:, :, 0]
    plt.imshow(im2)
    plt.show()
```

In the tutorial we'll be creating two scripts. First, image processing opreations on an image from Jurassic Park. Then counting the number of objects in a Tetris image

```python
# images are multidimensional Numpy arrays with shape height x width x channels
image = cv2.imread("jp.png")
(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))
```

```python
# display the image
cv2.imshow("Image", image)
cv2.waitKey(1000)
```

All images consists of pixels which are the raw building blocks of images. Images are made of pixels in a grid. A 640 x 480 image has 640 columns and 480 rows.<br><br>
Grayscale images have one number 0 to 255<br><br>
OpenCV images are in BGR order

```python
# extract channels from a pixel
(B, G, R) = image[100, 50]
print("R={}, G={}, B={}".format(R, G, B))
```

We can extract ROIs (regions of interest through slicing)

```python
# manually extract a face
roi = image[60:160, 320:420]
cv2.imshow("ROI", roi)
cv2.waitKey(0)
```

Resizing images can be important. You might want something to be larger on thee screen. Image processing is faster on smaller images. In deep learning, we often resize images, ignoring aspect ratio, so that the volume fits into a network which requires an image be square and of a certain dimension

```python
# regular resize
r = 300.0 / w
dim = (300, int(h * r))
resized = cv2.resize(image, dim)
cv2.imshow("Aspect Ratio Resize", resized)
cv2.waitKey(0)

# resize with imutils
resized = imutils.resize(image, width=300)
```

We can also rotate images

```python
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))

# with imutils
rotated = imutils.rotate(image, -45)

# with imutils and doesn't clip it after rotation
rotated = imutils.rotate_bound(image, 45)
```

## Counting Objects

- learning how to convert images to grayscale with OpenCV
- performing edge detection
- thresholding a grayscale image
- finding, counting, and drawing countours
- conduction erosion and dilation
- masking an image


```python
image = cv2.imread('tetris_blocks.png')
display_BGR(image)
```

We need to convert to grayscale because we will be thresholding and detecting edges in the image shortly

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

### Edge detection
useful for finding boundaries of objects in an image<br>

Canny function takes in 
- img
- minval
- maxval
- aperture_size

```python
edged = cv2.Canny(gray, 30, 150)
plt.imshow(edged, cmap='gray')
```

### Thresholding

thresholding is an important intermediary step for image processing pipelines. Thresholding can help us to remove lighter or darker regions and contours of images <br>
below we are setting pixels greater than 225 to 0 (black) and alll other pixels less than to 255

```python
thresh = cv2.threshold(gray, 225, 255, cv2.THRESH_BINARY_INV)[1]
plt.imshow(thresh, cmap='gray')
```

### Detecting and drawing contours

We have 6 shape contours. Llets find and draw their outlines via code

```python
# Draw contours
cnts = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, 
                       cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
output = image.copy()

for c in cnts:
    cv2.drawContours(output, [c], -1, (240, 0, 159), 3)
    display_BGR(output)
    
# Overlay text onto contours
text = "I found {} objects!".format(len(cnts))
cv2.putText(output, text, (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7,
           (240, 0, 159), 2)
```

### Erosions and Dilations

used to reduce noise in binary images. Reduce the size of foreground objects by eroding away pixels given a number of iterations

```python
mask = thresh.copy()
mask = cv2.erode(mask, None, iterations=5)
plt.figure()
plt.imshow(thresh, cmap='gray')
plt.figure()
plt.imshow(mask, cmap='gray')
```

similarly, we can foreground regions in the mask
